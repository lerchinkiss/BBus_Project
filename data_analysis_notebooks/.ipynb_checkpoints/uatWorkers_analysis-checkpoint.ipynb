```python
import pandas as pd
import os
from openpyxl import load_workbook
from openpyxl.utils import get_column_letter
```

```python
file_path = '../datasets/uatWorkers.xlsx' 
uatWorkers = pd.read_excel(file_path)
```

```python
#Посмотрим на количество данных в Excel файле
print("Размер таблицы (строки, столбцы):", uatWorkers.shape)
```

```python
print("Названия колонок:", uatWorkers.columns.tolist())
```

```python
#Посмотрим на значения данных в bbOrders
uatWorkers.head(20)
```

```python
#Оставим в uatWorkers нужные колонки для работы
columns_to_keep = [
    'Ref', 'IsFolder', 'DeletionMark', 'Parent', 'Code', 'Description',
    'ФизЛицо', 'Организация', 'ДатаПриема', 'ДатаУвольнения', 'Комментарий', 
    'ЗапрещенаПеревозкаДетей', 'ДатаОкончанияЗапретаНаПеревозкуДетей'
]
```

```python
# Проверка наличия колонок
missing_columns = [col for col in columns_to_keep if col not in uatWorkers.columns]

if missing_columns:
    print("В датафрейме отсутствуют следующие колонки:")
    for col in missing_columns:
        print(" -", col)

# Фильтрация по доступным колонкам
available_columns = [col for col in columns_to_keep if col in uatWorkers.columns]
uatWorkers_filtered = uatWorkers[available_columns]

# Проверка, что в bbOrders_filtered остались только нужные колонки
if set(uatWorkers_filtered.columns) == set(columns_to_keep):
    print("В датафрейме остались только нужные колонки.")
else:
    extra_columns = set(uatWorkers_filtered.columns) - set(columns_to_keep)
    if extra_columns:
        print("Остались лишние колонки в отфильтрованном датафрейме:")
        for col in extra_columns:
            print(" -", col)

# Вывод результата
print("\nРазмер отфильтрованного датафрейма:", uatWorkers_filtered.shape)
uatWorkers_filtered.head()
```

```python
# Замена всех NaN
uatWorkers_filtered = uatWorkers_filtered.fillna("Not Information")
uatWorkers_filtered.head()
```

```python
#Исходя из того, что в таблице присутствует достаточное количество значений "00000000-0000-0000-0000-000000000000", заменим это значение на Not information
uatWorkers_filtered = uatWorkers_filtered.replace('00000000-0000-0000-0000-000000000000', 'Not Information')
uatWorkers_filtered = uatWorkers_filtered.replace(0.0, 'Not Information')
uatWorkers_filtered.head()
```

```python
#Посмотрим на уникальные значения каждой колонки
for col in uatWorkers_filtered.columns:
    print(f"\nУникальные значения в колонке '{col}': ({uatWorkers_filtered[col].nunique()} всего)")
    print(uatWorkers_filtered[col].value_counts().head(10))
```

```python
# Убедимся, что папка существует
output_dir = '../filtered_datasets'
os.makedirs(output_dir, exist_ok=True)

# Путь к файлу
output_path = os.path.join(output_dir, 'uatWorkers_filtered.xlsx')

# Сохраняем в Excel
uatWorkers_filtered.to_excel(output_path, index=False)

# Открываем файл и настраиваем ширину колонок, из-за влияния на отображение данных
wb = load_workbook(output_path)
ws = wb.active

for i, column_cells in enumerate(ws.columns, 1):
    max_length = max(len(str(cell.value)) if cell.value is not None else 0 for cell in column_cells)
    adjusted_width = max_length + 2
    ws.column_dimensions[get_column_letter(i)].width = adjusted_width

# Сохраняем изменения
wb.save(output_path)
print(f"Файл сохранен: {output_path}")
```
