```python
import pandas as pd
import os
```

```python
file_path = '../datasets/bbTransportLocations.xlsx' 
bbTransportLocations = pd.read_excel(file_path)
```

```python
#Посмотрим на количество данных в Excel файле
print("Размер таблицы (строки, столбцы):", bbTransportLocations.shape)
```

```python
print("Названия колонок:", bbTransportLocations.columns.tolist())
```

```python
#Посмотрим на значения данных в bbTransportLocations
bbTransportLocations.head(20)
```

```python
columns_to_keep = [
    'Ref', 'DeletionMark', 'Parent', 'Code', 'Description','Адрес'
]
```

```python
# Проверка наличия колонок
missing_columns = [col for col in columns_to_keep if col not in bbTransportLocations.columns]

if missing_columns:
    print("В датафрейме отсутствуют следующие колонки:")
    for col in missing_columns:
        print(" -", col)

# Фильтрация по доступным колонкам
available_columns = [col for col in columns_to_keep if col in bbTransportLocations.columns]
bbTransportLocations_filtered = bbTransportLocations[available_columns]

# Проверка, что в bbTransportLocations_filtered остались только нужные колонки
if set(bbTransportLocations_filtered.columns) == set(columns_to_keep):
    print("В датафрейме остались только нужные колонки.")
else:
    extra_columns = set(bbTransportLocations_filtered.columns) - set(columns_to_keep)
    if extra_columns:
        print("Остались лишние колонки в отфильтрованном датафрейме:")
        for col in extra_columns:
            print(" -", col)

# Вывод результата
print("\nРазмер отфильтрованного датафрейма:", bbTransportLocations_filtered.shape)
bbTransportLocations_filtered.head()
```

```python
# Замена всех NaN
bbTransportLocations_filtered = bbTransportLocations_filtered.fillna("Not Information")
bbTransportLocations_filtered.head()
```

```python
#Исходя из того, что в таблице присутствует достаточное количество значений "00000000-0000-0000-0000-000000000000", заменим это значение на Not information
bbTransportLocations_filtered = bbTransportLocations_filtered.replace('00000000-0000-0000-0000-000000000000', 'Not Information')
bbTransportLocations_filtered.head()
```

```python
#Посмотрим на уникальные значения каждой колонки
for col in bbTransportLocations_filtered.columns:
    print(f"\nУникальные значения в колонке '{col}': ({bbTransportLocations_filtered[col].nunique()} всего)")
    print(bbTransportLocations_filtered[col].value_counts().head(10))
```

```python
# Убедимся, что папка существует
output_dir = '../filtered_datasets'
os.makedirs(output_dir, exist_ok=True)

# Путь к файлу
output_path = os.path.join(output_dir, 'bbTransportLocations_filtered.xlsx')

bbTransportLocations_filtered.to_excel(output_path, index=False)
print(f"Файл успешно сохранен: {output_path}")
```
