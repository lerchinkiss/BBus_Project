```python
import pandas as pd
import os
from openpyxl import load_workbook
from openpyxl.utils import get_column_letter
```

```python
file_path = '../datasets/uatModelsTS.xlsx' 
uatModelsTS = pd.read_excel(file_path)
```

```python
#Посмотрим на количество данных в Excel файле
print("Размер таблицы (строки, столбцы):", uatModelsTS.shape)
```

```python
print("Названия колонок:", uatModelsTS.columns.tolist())
```

```python
#Посмотрим на значения данных в bbOrders
uatModelsTS.head(20)
```

```python
#Оставим в uatModelsTS нужные колонки для работы
columns_to_keep = [
    'Ref', 'IsFolder', 'DeletionMark', 'Code', 'Description', 
    'ТипДвигателя', 'МодельДвигателя', 'МощностьДвигателяКВТ', 'МощностьДвигателяЛС',
    'РабочийОбъемДвигателя', 'КоличествоОсей', 'КоличествоКолес', 'МаксимальнаяСкорость', 
    'СобственныйВес', 'Грузоподъемность', 'ТипТС', 'ОсновноеТопливо', 'ДополнительноеТопливо', 
    'ВсегоМест', 'ВидМоделиТС'
]
```

```python
# Проверка наличия колонок
missing_columns = [col for col in columns_to_keep if col not in uatModelsTS.columns]

if missing_columns:
    print("В датафрейме отсутствуют следующие колонки:")
    for col in missing_columns:
        print(" -", col)

# Фильтрация по доступным колонкам
available_columns = [col for col in columns_to_keep if col in uatModelsTS.columns]
uatModelsTS_filtered = uatModelsTS[available_columns]

# Проверка, что в bbOrders_filtered остались только нужные колонки
if set(uatModelsTS_filtered.columns) == set(columns_to_keep):
    print("В датафрейме остались только нужные колонки.")
else:
    extra_columns = set(uatModelsTS_filtered.columns) - set(columns_to_keep)
    if extra_columns:
        print("Остались лишние колонки в отфильтрованном датафрейме:")
        for col in extra_columns:
            print(" -", col)

# Вывод результата
print("\nРазмер отфильтрованного датафрейма:", uatModelsTS_filtered.shape)
uatModelsTS_filtered.head()
```

```python
# Замена всех NaN
uatModelsTS_filtered = uatModelsTS_filtered.fillna("Not Information")
uatModelsTS_filtered.head()
```

```python
#Исходя из того, что в таблице присутствует достаточное количество значений "00000000-0000-0000-0000-000000000000" и 0.0, заменим это значение на Not information
uatModelsTS_filtered = uatModelsTS_filtered.replace('00000000-0000-0000-0000-000000000000', 'Not Information')
uatModelsTS_filtered = uatModelsTS_filtered.replace(0.0, 'Not Information')
uatModelsTS_filtered.head()
```

```python
#Посмотрим на уникальные значения каждой колонки
for col in uatModelsTS_filtered.columns:
    print(f"\nУникальные значения в колонке '{col}': ({uatModelsTS_filtered[col].nunique()} всего)")
    print(uatModelsTS_filtered[col].value_counts().head(10))
```

```python
# Фильтрация по условию
mask = uatModelsTS_filtered['ВсегоМест'] == 'Not Information'

# Получение списка значений Description, соответствующих условию
descriptions_with_missing_seats = uatModelsTS_filtered.loc[mask, 'Description'].unique().tolist()

# Вывод
print("Модели с отсутствующим количеством мест:")
for desc in descriptions_with_missing_seats:
    print(f" - {desc}")
```

```python
#Создадим словарь с транспортом и количеством мест
seat_mapping = {
    "MAN LION'S COACH": 55,
    "Yutong (49+1)": 49,
    "MAN (49+1)": 49,
    "Yutong (52+1)": 52,
    "Хайгер (51+1)": 52,
    "Neoplan": 55,
    "King Long (50+1)": 50,
    "Ютонг (51+1)": 51,
    "Higer (49+1)": 49,
    "King Long": 53,
    "King Long (51+1)": 51,
    "Ютонг 51": 51,
    "Scania A80": 55,
    "Mercedes (51+1)": 51,
    "Автобус грузовой": 20, 
    "Голден Драгон": 53,
    "King Long (53+1)": 53,
    "неоплан": 55,
    "ЛиАЗ 529115": 29,
    "ПАЗ 320405-04": 22,
    "ПАЗ 320412-04": 22,
    "ПАЗ 320412-05": 22,
    "MAN": 55,
    "ЛиАЗ 429260": 23,
    "Mercedes Vito": 8,
    "MB Sprinter 19": 19,
    "Mersedes Sprinter(19)": 19,
    "В Mercedes-Benz Sprinter (19+1)": 19,
    "MB Sprinter 20": 20,
    "MB Sprinter16": 16,
    "MB Vito": 8,
    "Пежо Боксер": 8,
    "CITROEN": 18,
    "Skoda Octavia": 4,
    "Toyota Camry": 4,
    "Тойота Камри": 4,
    "Тойота Ленд Крузер 200": 4,
    "Шкода Октавия": 4,
    "Скания": 55,
    "Neoplan 49+1": 49,
    "Higer (51)": 51,
    "Мерседес Спринтер (19+1)": 19,
    "Mercedes V class": 8,
    "6122": 53,
    "Тойота RAV4": 4
}
```

```python
#Заменим количество мест в ВсегоМест подходящие под каждый транспорт
def fill_seats_from_mapping(row):
    if row['ВсегоМест'] == "Not Information":
        return seat_mapping.get(row['Description'], row['ВсегоМест'])
    return row['ВсегоМест']

# Применение к датафрейму
uatModelsTS_filtered['ВсегоМест'] = uatModelsTS_filtered.apply(fill_seats_from_mapping, axis=1)
```

```python
# Убедимся, что папка существует
output_dir = '../filtered_datasets'
os.makedirs(output_dir, exist_ok=True)

# Путь к файлу
output_path = os.path.join(output_dir, 'uatModelsTS_filtered.xlsx')

# Сохраняем в Excel
uatModelsTS_filtered.to_excel(output_path, index=False)

# Открываем файл и настраиваем ширину колонок, из-за влияния на отображение данных
wb = load_workbook(output_path)
ws = wb.active

for i, column_cells in enumerate(ws.columns, 1):
    max_length = max(len(str(cell.value)) if cell.value is not None else 0 for cell in column_cells)
    adjusted_width = max_length + 2
    ws.column_dimensions[get_column_letter(i)].width = adjusted_width

# Сохраняем изменения
wb.save(output_path)
print(f"Файл сохранен: {output_path}")
```
