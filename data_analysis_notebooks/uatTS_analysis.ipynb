```python
import pandas as pd
import os
from openpyxl import load_workbook
from openpyxl.utils import get_column_letter
```

```python
file_path = '../datasets/uatTS.xlsx' 
uatTS = pd.read_excel(file_path)
```

```python
print("Названия колонок:", uatTS.columns.tolist())
```

```python
#Посмотрим на значения данных в uatTS
uatTS.head(20)
```

```python
#Оставим в uatTS нужные колонки для работы
columns_to_keep = [
    'Ref', 'IsFolder', 'DeletionMark', 'Parent', 'Description', 'ОграниченияДляЗаказов', 
    'Параметры', 'Гараж', 'ГаражныйНомер', 'ГодВыпуска', 'ГосударственныйНомер',
    'ДатаВводаВЭксплуатацию', 'ДатаВыбытия', 'ДатаПостановкиНаУчетГИБДД', 
    'Комментарий', 'Модель', 'МодельДвигателя', 'МощностьДвигателяКВт', 'МощностьДвигателяЛС', 
    'НачальныйПробег', 'ТипТС', 'ЦветКузова',  'TTID', 'Исполнитель'
]
```

```python
# Проверка наличия колонок
missing_columns = [col for col in columns_to_keep if col not in uatTS.columns]

if missing_columns:
    print("В датафрейме отсутствуют следующие колонки:")
    for col in missing_columns:
        print(" -", col)

# Фильтрация по доступным колонкам
available_columns = [col for col in columns_to_keep if col in uatTS.columns]
uatTS_filtered = uatTS[available_columns]

# Проверка, что в uatTS_filtered остались только нужные колонки
if set(uatTS_filtered.columns) == set(columns_to_keep):
    print("В датафрейме остались только нужные колонки.")
else:
    extra_columns = set(uatTS_filtered.columns) - set(columns_to_keep)
    if extra_columns:
        print("Остались лишние колонки в отфильтрованном датафрейме:")
        for col in extra_columns:
            print(" -", col)

# Вывод результата
print("\nРазмер отфильтрованного датафрейма:", uatTS_filtered.shape)
uatTS_filtered.head()
```

```python
# Замена всех NaN
uatTS_filtered = uatTS_filtered.fillna("Not Information")
uatTS_filtered.head()
```

```python
#Исходя из того, что в таблице присутствует достаточное количество значений "00000000-0000-0000-0000-000000000000", заменим это значение на Not information
uatTS_filtered = uatTS_filtered.replace('00000000-0000-0000-0000-000000000000', 'Not Information')
uatTS_filtered = uatTS_filtered.replace(0.0, 'Not Information')
uatTS_filtered.head()
```
