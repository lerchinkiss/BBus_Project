```python
import pandas as pd
```

```python
file_path = '../datasets/bbOrders.xlsx' 
bbOrders = pd.read_excel(file_path)
```

```python
#Посмотрим на количество данных в Excel файле
print("Размер таблицы (строки, столбцы):", bbOrders.shape)
```

```python
print("Названия колонок:", bbOrders.columns.tolist())
```

```python
#Посмотрим на значения данных в bbOrders
bbOrders.head(20)
```

```python
#Оставим в bbOrders нужные колонки для работы
columns_to_keep = [
    'Ref', 'Date', 'Number', 'Заказчик', 'ВремяНачала', 'ПланируемоеВремяОкончания',
    'Загрузка', 'Разгрузка', 'ТС', 'Комментарии','ПланируемоеВремяРаботы', 
    'ФактическоеВремяРаботы', 'ПланируемоеВремяПодачи', 'ФактическоеВремяПодачи', 
    'КонтактноеЛицо', 'ЦенаЗаЧас', 'Организация', 'Ответсвенный',
    'РасчетнаяСтоимость', 'ФактическаяСтоимость', 'Модель', 'КоличествоПассажиров',
    'ТипТС', 'Водитель', 'Тариф', 'Перекат', 'Табличка', 'Исполнитель', 'СобственноеТС',
    'ВнутреннийНомерУЗаказчика', 'Оплачено', 'ДатаОплаты', 'ЖдемОплатуДо',
    'ОтправленоСМСВодителю', 'ОтправленоСМСЗаказчику', 'ЗагрузкаПункт', 'ЗагрузкаАдрес',
    'РазгрузкаПункт', 'РазгрузкаАдрес', 'НомерПП', 'Счет', 'Акт', 'Создатель', 'Водитель2',
    'ЧасыВодителю', 'ЧасыВодителю2', 'ОтправлятьСМСКонтактномуЛицу', 'НалПолучено',
    'ТребуетсяВторойВодитель', 'ЗарубежнаяПоездка', 'НомерРодительскогоЗаказа',
    'ВидПредоплаты2', 'НалПолучено2', 'ТипЗаказа', 'СтатусЗаказа', 'ЦенаЗаЧасИсполнителя',
    'КомментарийКМаршруту', 'ТребуетсяТуалет', 'Многодневный', 'Аннулирован',
    'ДатаАннуляции', 'ПричинаАннуляции', 'УтвержденДиспетчером', 'КомментарииДляВодителя',
    'ДанныеОтданы', 'Акция', 'НомерППКопия', 'КомментарииПоЗавершению', 'Багаж',
    'Уведомление', 'Задачи', 'ДопУслуги', 'ПромежуточныеТочки'
]
```

```python
# Проверка наличия колонок
missing_columns = [col for col in columns_to_keep if col not in bbOrders.columns]

if missing_columns:
    print("В датафрейме отсутствуют следующие колонки:")
    for col in missing_columns:
        print(" -", col)

# Фильтрация по доступным колонкам
available_columns = [col for col in columns_to_keep if col in bbOrders.columns]
bbOrders_filtered = bbOrders[available_columns]

# Проверка, что в bbOrders_filtered остались только нужные колонки
if set(bbOrders_filtered.columns) == set(columns_to_keep):
    print("В датафрейме остались только нужные колонки.")
else:
    extra_columns = set(bbOrders_filtered.columns) - set(columns_to_keep)
    if extra_columns:
        print("Остались лишние колонки в отфильтрованном датафрейме:")
        for col in extra_columns:
            print(" -", col)

# Вывод результата
print("\nРазмер отфильтрованного датафрейма:", bbOrders_filtered.shape)
bbOrders_filtered.head()
```

```python
# Замена всех NaN
bbOrders_filtered = bbOrders_filtered.fillna("Not Information")
bbOrders_filtered.head()
```

```python
print("\nТипы данных по колонкам:")
print(bbOrders_filtered.dtypes.value_counts())
print(bbOrders_filtered.dtypes)
```

```python
#Посмотрим на уникальные значения каждой колонки
for col in bbOrders_filtered.columns:
    print(f"\nУникальные значения в колонке '{col}': ({bbOrders_filtered[col].nunique()} всего)")
    print(bbOrders_filtered[col].value_counts().head(10))
```

```python
#Исходя из того, что в таблице присутствует достаточное количество значений "00000000-0000-0000-0000-000000000000", заменим это значение на Not information
bbOrders_filtered = bbOrders_filtered.replace('00000000-0000-0000-0000-000000000000', 'Not Information')
bbOrders_filtered.head()
```

```python
numeric_cols = ['ЦенаЗаЧас', 'ФактическаяСтоимость', 'РасчетнаяСтоимость']
for col in numeric_cols:
    try:
        bbOrders_filtered[col] = pd.to_numeric(bbOrders_filtered[col], errors='coerce')
        print(f"\nКолонка: {col}")
        print(bbOrders_filtered[col].describe())
    except Exception as e:
        print(f"Ошибка при обработке {col}: {e}")
```

```python
#Сохраним отфильтрованный датафрейм и поместим в папку
import os

# Убедимся, что папка существует
output_dir = '../filtered_datasets'
os.makedirs(output_dir, exist_ok=True)

# Путь к файлу
output_path = os.path.join(output_dir, 'bbOrders_filtered.xlsx')

bbOrders_filtered.to_excel(output_path, index=False)
print(f"Файл успешно сохранен: {output_path}")
```
