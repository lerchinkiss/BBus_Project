```python
import pandas as pd
import os
from openpyxl import load_workbook
from openpyxl.utils import get_column_letter
```

```python
file_path = '../datasets/Users.xlsx' 
Users = pd.read_excel(file_path)
```

```python
#Посмотрим на количество данных в Excel файле
print("Размер таблицы (строки, столбцы):", Users.shape)
```

```python
print("Названия колонок:", Users.columns.tolist())
```

```python
#Посмотрим на значения данных в bbOrders
Users.head(20)
```

```python
#Оставим в Users нужные колонки для работы
columns_to_keep = [
    'Ref', 'IsFolder', 'DeletionMark', 'Parent', 'Code', 'Description', 
    'ФизЛицо', 'Контрагент', 'id_pyrus', 'ЛогинПирус', 'Роль'
]
```

```python
# Проверка наличия колонок
missing_columns = [col for col in columns_to_keep if col not in Users.columns]

if missing_columns:
    print("В датафрейме отсутствуют следующие колонки:")
    for col in missing_columns:
        print(" -", col)

# Фильтрация по доступным колонкам
available_columns = [col for col in columns_to_keep if col in Users.columns]
Users_filtered = Users[available_columns]

# Проверка, что в bbOrders_filtered остались только нужные колонки
if set(Users_filtered.columns) == set(columns_to_keep):
    print("В датафрейме остались только нужные колонки.")
else:
    extra_columns = set(Users_filtered.columns) - set(columns_to_keep)
    if extra_columns:
        print("Остались лишние колонки в отфильтрованном датафрейме:")
        for col in extra_columns:
            print(" -", col)

# Вывод результата
print("\nРазмер отфильтрованного датафрейма:", Users_filtered.shape)
Users_filtered.head()
```

```python
# Замена всех NaN
Users_filtered = Users_filtered.fillna("Not Information")
Users_filtered.head()
```

```python
#Исходя из того, что в таблице присутствует достаточное количество значений "00000000-0000-0000-0000-000000000000" и 0.0, заменим это значение на Not information
Users_filtered = Users_filtered.replace('00000000-0000-0000-0000-000000000000', 'Not Information')
Users_filtered = Users_filtered.replace(0.0, 'Not Information')
Users_filtered.head()
```

```python
# Убедимся, что папка существует
output_dir = '../filtered_datasets'
os.makedirs(output_dir, exist_ok=True)

# Путь к файлу
output_path = os.path.join(output_dir, 'Users_filtered.xlsx')

# Сохраняем в Excel
Users_filtered.to_excel(output_path, index=False)

# Открываем файл и настраиваем ширину колонок, из-за влияния на отображение данных
wb = load_workbook(output_path)
ws = wb.active

for i, column_cells in enumerate(ws.columns, 1):
    max_length = max(len(str(cell.value)) if cell.value is not None else 0 for cell in column_cells)
    adjusted_width = max_length + 2
    ws.column_dimensions[get_column_letter(i)].width = adjusted_width

# Сохраняем изменения
wb.save(output_path)
print(f"Файл сохранен: {output_path}")
```
