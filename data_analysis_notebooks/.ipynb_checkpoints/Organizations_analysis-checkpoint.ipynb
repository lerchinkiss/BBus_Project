```python
import pandas as pd
import os
from openpyxl import load_workbook
from openpyxl.utils import get_column_letter
```

```python
file_path = '../datasets/Organizations.xlsx' 
Organizations = pd.read_excel(file_path)
```

```python
#Посмотрим на количество данных в Excel файле
print("Размер таблицы (строки, столбцы):", Organizations.shape)
```

```python
print("Названия колонок:", Organizations.columns.tolist())
```

```python
#Посмотрим на значения данных в Organizations
Organizations.head(20)
```

```python
#Оставим в Organizations нужные колонки для работы
columns_to_keep = [
    'Ref', 'DeletionMark', 'Code', 'Description', 'Префикс', 'НаименованиеПолное','ЮрФизЛицо'
]
```

```python
# Проверка наличия колонок
missing_columns = [col for col in columns_to_keep if col not in Organizations.columns]

if missing_columns:
    print("В датафрейме отсутствуют следующие колонки:")
    for col in missing_columns:
        print(" -", col)

# Фильтрация по доступным колонкам
available_columns = [col for col in columns_to_keep if col in Organizations.columns]
Organizations_filtered = Organizations[available_columns]

# Проверка, что в bbOrders_filtered остались только нужные колонки
if set(Organizations_filtered.columns) == set(columns_to_keep):
    print("В датафрейме остались только нужные колонки.")
else:
    extra_columns = set(Organizations_filtered.columns) - set(columns_to_keep)
    if extra_columns:
        print("Остались лишние колонки в отфильтрованном датафрейме:")
        for col in extra_columns:
            print(" -", col)

# Вывод результата
print("\nРазмер отфильтрованного датафрейма:", Organizations_filtered.shape)
Organizations_filtered.head()
```

```python
# Убедимся, что папка существует
output_dir = '../filtered_datasets'
os.makedirs(output_dir, exist_ok=True)

# Путь к файлу
output_path = os.path.join(output_dir, 'Organizations_filtered.xlsx')

# Сохраняем в Excel
Organizations_filtered.to_excel(output_path, index=False)

# Открываем файл и настраиваем ширину колонок, из-за влияния на отображение данных
wb = load_workbook(output_path)
ws = wb.active

for i, column_cells in enumerate(ws.columns, 1):
    max_length = max(len(str(cell.value)) if cell.value is not None else 0 for cell in column_cells)
    adjusted_width = max_length + 2
    ws.column_dimensions[get_column_letter(i)].width = adjusted_width

# Сохраняем изменения
wb.save(output_path)
print(f"Файл сохранен: {output_path}")
```
