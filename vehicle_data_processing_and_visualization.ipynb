```python
import os
import pandas as pd
```

```python
datasets_folder = "datasets" 
file_path = os.path.join(datasets_folder, "bus_details.csv")  
```

```python
#Загрузка датасета и вывод
bus_details = pd.read_csv(file_path)
bus_details
```

```python
print("\nРазмеры датасета (строки, столбцы):")
print(bus_details.shape)
```

```python
print("\nКоличество пропущенных значений в каждом столбце:")
print(bus_details.isnull().sum())
```

```python
print("\nОбщая информация о датасете:")
print(bus_details.info())
```

```python
#Замена в перечисленных колонка отсутствующих значений
columns_to_update = ['Багажник', 'Ремни', 'Монитор', 'Микрофон', 'Розетки', 'Опция', 'Удобства']
bus_details_upgrade = bus_details.copy()
bus_details_upgrade[columns_to_update] = bus_details_upgrade[columns_to_update].fillna("отсутствует информация")
bus_details_upgrade
```

```python
#Замена пустых значений в колонках - Цвет салона, Цвет, Привод
bus_details_upgrade['Цвет салона'] = bus_details_upgrade['Цвет салона'].fillna("серый")
bus_details_upgrade['Цвет'] = bus_details_upgrade['Цвет'].fillna("белый")
bus_details_upgrade['Привод'] = bus_details_upgrade['Привод'].fillna("задний")
bus_details_upgrade
```

```python
print("\nПосле обработки пропущенных значений:")
print(bus_details_upgrade.isnull().sum())
```

```python
#Определение правил преобразования
mappings = {
    "Услуга": {"аренда автобуса": 1, "аренда микроавтобуса": 2, "аренда минивэна": 3, "аренда авто": 4},
    "Багажник": {"есть": 1, "отсутствует информация": 0},
    "Ремни": {"есть": 1, "отсутствует информация": 0},
    "Салон": {"комби": 1, "тканевый": 2, "кожаный": 3, "кожа/ткань": 1, "Эко-кожа": 4},
    "Климат": {"кондиционер": 1, "климат-контроль": 2},
    "Цвет салона": {
        "синеоранжевый/серый": 1,
        "синий-оранжевый": 2,
        "бордовый-черный": 3,
        "чёрно-оранжево-бордовый": 4,
        "серый": 5,
        "серый-оранжевый": 6,
        "синий": 7,
    },
    "Монитор": {"1 LED": 1, "2 LED": 2, "отсутствует информация": 0},
    "Микрофон": {"есть": 1, "отсутствует информация": 0},
    "Розетки": {"USB, 220 B": 1, "отсутствует информация": 0, "USB": 2},
    "Опция": {"холодильник": 1, "отсутствует информация": 0},
    "Удобства": {"туалет": 1, "отсутствует информация": 0},
    "Цвет": {"белый": 1, "Чёрный": 0, "Чёрный/белый": 3},
    "Привод": {"задний": 1, "передний": 2, "полный": 3, "перед./полн.": 4},
}
#Преобразование значений в числовые
for column, mapping in mappings.items():
    if column in bus_details_upgrade.columns:
        bus_details_upgrade[column] = bus_details_upgrade[column].map(mapping)
if "Min аренда" in bus_details_upgrade.columns:
    bus_details_upgrade["Min аренда"] = bus_details_upgrade["Min аренда"].str.extract(r'(\d+)').astype(int)

bus_details_upgrade
```

```python
#Замена пустых значений в колонках - Цвет салона, Цвет, Привод
bus_details_upgrade['Цвет салона'] = bus_details_upgrade['Цвет салона'].fillna("серый")
bus_details_upgrade['Цвет'] = bus_details_upgrade['Цвет'].fillna("белый")
bus_details_upgrade['Привод'] = bus_details_upgrade['Привод'].fillna("задний")
bus_details_upgrade
```

```python
print("\nПосле обработки пропущенных значений:")
print(bus_details_upgrade.isnull().sum())
```
